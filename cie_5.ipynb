```python
import spacy
from nltk import ngrams
nlp = spacy.load("en_core_web_sm")
doc = nlp("AI is simulation of human intelligence by machibne. it includes expects system NLP speech recognition and machine vision")
bigrams = ngrams(doc,2)
print(list(bigrams))
```
